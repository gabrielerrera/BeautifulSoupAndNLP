```
import json
import os
import requests
import string
from bs4 import BeautifulSoup
```

```
BASE_URL = "http://www.fatec.edu.br/category/aluno/"
DATA_DIR = "fatec"
```

```
response = requests.get(BASE_URL)
```

```
bs = BeautifulSoup(response.text)
```

```
vagas = bs.find_all(class_="modal")
```

```
class FatecEstagio:
  def __init__(self,bs_vaga=None):

    if bs_vaga is not None:
      self.nome = bs_vaga.find("strong").text.translate(str.maketrans('', '', string.punctuation))
      self.conteudo = bs_vaga.find("p").text

  def save_tojson(p, savedir, name=None):
    if name is None:
      name = f"{p.nome}.json"
    path = os.path.join(savedir, name)
    with open(path, "w") as jsonfile:
      json.dump(p.__dict__, jsonfile)

  def load_fromjson(filepath):
    with open(filepath, "r") as jsonfile:
      data = json.load(jsonfile)
      p = FatecEstagio()
      p.nome = data["name"]
      p.conteudo = data["text"]
      return p
```

```
jobs = [FatecEstagio(art) for art in vagas]
```

```
for p in jobs:
  FatecEstagio.save_tojson(p, DATA_DIR)
```
